- Задание


При фиксированном seed=42 поэкспериментируйте с параметрами алгоритма differential_evolution: strategy, popsize, mutation.


Постройте графики количества итераций (nit) оптимизации функции ackley от значения параметра.


```python
import numpy as np
from scipy.optimize import differential_evolution
import matplotlib.pyplot as plt
from scipy.optimize import minimize



%matplotlib inline
```

```python
def ackley(x):
    arg1 = -0.2 * np.sqrt(0.5 * (x[0] ** 2 + x[1] ** 2))
    arg2 = 0.5 * (np.cos(2. * np.pi * x[0]) + np.cos(2. * np.pi * x[1]))
    return -20. * np.exp(arg1) - np.exp(arg2) + 20. + np.e

bounds = [(-10, 10), (-10, 10)]

result = differential_evolution(ackley, bounds, seed = 42)
result
```

```python
res  =  [] 
param  =  [] 

for  i  in range(1, 100, 10):     
    res.append(differential_evolution(ackley,  bounds,  popsize = i,  seed = 42 )['nit']) 
    param.append(i) 

    
    
plt.plot(param,  res, color = 'r')
plt.title('Зависимость количества итераций от параметра popsize')
plt.ylabel('Кол-во итераций') 
plt.xlabel('popsize') 
plt.show()
```

```python
res1  =  [] 
param1  =  [] 

for i in ['best1bin', 'best1exp', 'rand1exp', 'randtobest1exp', 'currenttobest1exp', 
          'best2exp','rand2exp','randtobest1bin', 'currenttobest1bin', 'best2bin', 'rand2bin', 'rand1bin']:
    res1.append(differential_evolution(ackley,  bounds,  strategy = i,  seed = 42 )['nit']) 
    param1.append(i) 

    
    
plt.plot(param1,  res1, color = 'b')
plt.title('Зависимость количества итераций от параметра strategy')
plt.ylabel('Кол-во итераций') 
plt.xlabel('strategy') 
plt.show()

```

```python
res2  =  [] 
param2  =  [] 

for i in range(0, 2, 1) :
    res2.append(differential_evolution(ackley,  bounds,  mutation = i,  seed = 42 )['nit']) 
    param2.append(i) 

    
    
plt.plot(param2,  res2, color = 'y')
plt.title('Зависимость количества итераций от параметра mutation')
plt.ylabel('Кол-во итераций') 
plt.xlabel('mutation') 
plt.show()
```
